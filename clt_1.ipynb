Studying Normal distribution as an approximation to the Binomial distribution is practical.


```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
f = math.factorial

def Binomial_PMF(n, k, p):
    return f(n)/(f(k) * f(n-k)) * p**k * (p -1)**(n-k)
```

```python
# Simulate 20 coin flips

coin_flips = 20
p = 0.5
exp = Binomial_PMF
outs = []

for i in range(coin_flips):
    e = exp(20, i, p)
    outs.append(e)
```

```python
outs[:10]
```

```python
plt.bar(range(20), np.abs(outs))
plt.xlabel('Numper of heads for each trail')
plt.ylabel('Binomial distribution at each trail')
plt.show()
```

```python
# simulate 3000 coin flips with getting the ratio of #heads/1000
H_freq_outs = []
for i in range(3000):
    c = np.random.choice(('H', 'T'), size=1000, p=(0.5, 0.5))
    H_freq = np.count_nonzero(c=='H')
    H_freq_outs.append(H_freq/1000)
```

```python
plt.hist(H_freq_outs)
plt.xlabel('Mean values')
plt.ylabel('Frequency')
plt.show()
```
